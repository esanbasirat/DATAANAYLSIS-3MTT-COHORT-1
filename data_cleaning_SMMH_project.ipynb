 PRINT ALL DATA SET IN THE CSV FILE.

```python
import pandas as pd
# Load the CSV file into a DataFrame
df = pd.read_csv('/content/SMAMH.csv')

```

```python
# Print the entire DataFrame
print(df)
```

DELETING ROW 252-482 BECAUSE THEY HAVE NO DATA

```python
# Delete rows based on the defined condition
# Delete rows 252 to 482
df.drop(df.index[251:482], inplace=True)
```

```python
print(df)
```

HANDLING MISSING DATA:Identify and handle missing values appropriately. This could involve dropping rows or columns with missing values.

```python
#Handle Missing Values
# Drop rows with missing values
df.dropna(inplace=True)

```

```python
# Or, impute missing values
# df.fillna(value, inplace=True)
```

```python
print(df)
```

 DELETING CERTAIN COLUMN

```python
# Delete a certain column (e.g., 'column_to_delete')
# Delete column 21 (index 20 since Python is 0-indexed)
df.drop(df.columns[20], axis=1, inplace=True)
print(df)
```
